```python
# %load C:/Users/evgen/Parsing/definitions/url_test.py
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
import re
import pypandoc
from nltk import sent_tokenize


import spacy
from spacy.tokenizer import Tokenizer
nlp = spacy.load('en_core_web_sm')

class url_test:    
    def __init__(self, myUrl):
        #----Will show the url
        self.myUrl = myUrl
        #----Will show the url string[0] as definition section from the page
        #----Will show the url string[1] if there is a table in definition section. In case it is will return "False"
        self.string = self.__parsing()
    def __parsing(self):
        
        def parsed_string():
            uClient = uReq(self.myUrl)
            page_html = uClient.read()
            uClient.close()
            page_soup = soup(page_html, "html.parser")
            TBData = page_soup.find("textarea",{"id": "wpTextbox1"})
            # Returning  list of strings as it was presented in the webpage
            return TBData.text.splitlines()
        def findDefinition(str_):
            #Deleting white spaces and "\r" characters
            whitespacesDict =	{
                "4": "\u2004",
                "5": "\u2005",
                "6": "\u2006",
                "\r": "\r"
                }
            #Cleanning empty lines
            for line in str_:
                if not line.strip():
                    str_.remove(line)
            #findin definition text section        
            flag=False
            triple_flag=False
            idx=0
            result =''
            for line in str_:
                if flag==True:
                    if re.search(r'^={2}[^=].*[^=]={2}$',line):
                        #----Here we stop scan  for definition area.
                        #----Definition area is presented after "==Definition==" sentence and 
                        #-----finishes before next "==Any word or sentence=="
                        break
                    if re.search(r'^={3}\s*.*\s*={3}$',line):
                        #-----Inside of definitioan text area we saw inner section whic starts with 
                        #----------"===Any word or sentence===", we skip on this sentence here.
                        triple_flag=True
                        idx+=1
                        continue
                    #------ Replacing any hyperlinks with double bracket
                    line = re.sub(r'\[{2}\w*?(\ \w+)+\:{2}', '[[', line)
                    #-------Cleaning whitespaces (unicode)
                    for value in whitespacesDict.values():
                        line = line.replace(value,"")
                    #-------Cleaning "\n" end of line
                    line = line.replace("\n"," ")
                    result+=line
                #-------Here "Definition" section found
                if re.search(r'^={2}\s*Definition\s*={2}$',line):
                    flag=True
            #-------Return definition text are in one string
            return result
        #-----Calling to the main function
        result = findDefinition(parsed_string())
        #-----Checks if there is a table in definition text area and return "True" if found
        table = re.search('{| class="sortable" border="1"',result)
        return [result,table is None]
```
